```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as ppl
import seaborn as sb

path = './data/Iris.csv'
df = pd.read_csv(path)
df.tail(-1)
df.drop('Id', axis=1, inplace=True)
```

```python
df.info()
df.describe().T
```

Такой график называется диаграмма рассеивания 
На которой можно увидеть три класса цветков

```python
y = df.iloc[0: 100, 4].values
y = np.where(y == 'Iris-setosa', -1, 1)
print(df.iloc[0: 100, [0, 2]])
x = df.iloc[0: 150, [0, 2]].values
ppl.scatter(x[:50, 0], x[:50, 1], color='red', marker='o', label='setosa')
ppl.scatter(x[50:100, 0], x[50:100, 1], color='blue', marker='o', label='versicolor')
ppl.scatter(x[100:150, 0], x[100:150, 1], color='green', marker='o',label='virginica')
ppl.xlabel('Длина чашелистика')
ppl.ylabel('Длина лепестка')
ppl.legend(loc='upper left')
ppl.show()
```
